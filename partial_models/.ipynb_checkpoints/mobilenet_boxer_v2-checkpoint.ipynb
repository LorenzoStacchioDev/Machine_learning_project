# Implement IOU for validation phase

```python
class Validation(Callback):
    def __init__(self, generator):
        self.generator = generator

    def on_epoch_end(self, epoch, logs):
        mse = 0
        intersections = 0
        unions = 0

        for i in range(len(self.generator)):
            batch_images, gt = self.generator[i]
            pred = self.model.predict_on_batch(batch_images)

            pred = np.maximum(pred, 0)
            diff_width = np.minimum(gt[:,0] + gt[:,2], pred[:,0] + pred[:,2]) - np.maximum(gt[:,0], pred[:,0])
            diff_height = np.minimum(gt[:,1] + gt[:,3], pred[:,1] + pred[:,3]) - np.maximum(gt[:,1], pred[:,1])
            intersection = np.maximum(diff_width, 0) * np.maximum(diff_height, 0)
            area_gt = gt[:,2] * gt[:,3]
            area_pred = pred[:,2] * pred[:,3]
            union = np.maximum(area_gt + area_pred - intersection, 0)

            intersections += np.sum(intersection * (union > 0))
            unions += np.sum(union)

        iou = np.round(intersections / (unions + epsilon()), 4)
        logs["val_iou"] = iou
        print("val_iou: {}".format(iou))
```

# Dataset

```python
train_df = pandas.read_csv("/data01/ML/dataset/FACE_CLASSIFIER/dataset_face_detection/train2.csv")
valid_df = pandas.read_csv("/data01/ML/dataset/FACE_CLASSIFIER/dataset_face_detection/val2.csv")
```

## Create generators

```python
train_datagen = ImageDataGenerator(
    preprocessing_function = preprocess_input)

train_generator = train_datagen.flow_from_dataframe(
        dataframe=train_df,
        directory=None,
        x_col="image_path",
        y_col=["x","y","x_2","y_2"],
        class_mode="raw",
        shuffle=True,
        color_mode = 'rgb',
        target_size=(IMG_SIZE, IMG_SIZE),
        batch_size=BATCH_SIZE)

valid_generator = train_datagen.flow_from_dataframe(
        dataframe=valid_df,
        directory=None,
        x_col="image_path",
        y_col=["x","y","x_2","y_2"],
        class_mode="raw",
        shuffle=True,
        color_mode = 'rgb',
        target_size=(IMG_SIZE, IMG_SIZE),
        batch_size=BATCH_SIZE)
```

```python
step_size_train = train_generator.samples/train_generator.batch_size
step_size_valid = valid_generator.samples/valid_generator.batch_size
```

# Fully Convolutional Model

```python
model =  MobileNetV2(include_top=False, weights='imagenet', input_shape=(224,224,3), alpha = 1.4)
```

```python
x = model.layers[-1].output
x = Conv2D(4,7, strides=(1,1), name="detector_conv_2", activation = "relu")(x)
x = Reshape((4,), name = "coords")(x)
model = Model(input = [model.input], outputs = x)
print(len(model.layers))
```

```python
model.summary()
```

# Training

```python
model = multi_gpu_model(model,gpus=2)
```

model.load_weights('model_val_iou_2.h5')

```python
validation_datagen = Validation(generator=valid_generator)

model.compile(optimizer=Adam(lr = 1e-3), loss="mse", metrics=[])
mcp_save = ModelCheckpoint("mobile_net_seconda_passata.h5", monitor="val_iou", verbose=1, save_best_only=True, save_weights_only=True, mode="max")
reduce_lr_loss = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=10, min_lr = 1e-7, verbose=1, mode='min')

history = model.fit_generator(generator=train_generator, epochs=30, steps_per_epoch=step_size_train, validation_data=valid_generator, validation_steps=step_size_valid, verbose=1, shuffle = True, callbacks=[validation_datagen,mcp_save,reduce_lr_loss])
```

```python
print(history.history.keys())
```

```python
loss = history.history['loss']
val_loss = history.history['val_loss']
val_IOU = history.history['val_iou']

plt.figure(figsize=(8, 8))
plt.subplot(2, 1, 1)
plt.plot(val_loss, label='Validation MSE')
plt.plot(loss, label='Training MSE')
plt.legend(loc='lower right')
plt.ylabel('MSE')
plt.ylim([min(plt.ylim()),100])
plt.title('Training and Validation MSE')

plt.subplot(2, 1, 2)
plt.plot(val_IOU, label='Validation IOU')
plt.legend(loc='lower right')
plt.ylabel('IOU')
plt.ylim([min(plt.ylim()),1])
plt.title('Training and Validation IOU')
#plt.save_figure("IOU_seconda_passata.png")
```
